# **üìò 1. O que √© o while?**

O **`while`** √© uma estrutura de **repeti√ß√£o** usada quando **n√£o sabemos** exatamente quantas vezes precisamos repetir um bloco de c√≥digo.

A repeti√ß√£o continua enquanto uma **condi√ß√£o for verdadeira**.

**üß© Sintaxe:**

```
while condi√ß√£o:
    # bloco de c√≥digo a ser repetido

```



* **Condi√ß√£o**: uma express√£o que √© avaliada como True ou False. Enquanto a condi√ß√£o for True, o loop continuar√° a ser executado.
* **Bloco de c√≥digo**: o conjunto de instru√ß√µes que ser√£o repetidas.
* Deve ser **indentado**.

# **üß† 2. Conceitos Importantes**

# **‚úÖ 2.1. Condi√ß√£o Booleana**

O while depende de uma **condi√ß√£o** que deve retornar **`True`** para continuar o loop.

````python
contador = 0
while contador < 5:
    print("Contador √©:", contador)
    contador += 1
````

**Explica√ß√£o do C√≥digo:**

* **Inicializa√ß√£o**: A vari√°vel contador √© inicializada com o valor 0.

* **Condi√ß√£o**: while contador < 5 verifica se o valor de contador √© menor que 5.

* **Bloco de C√≥digo:**

  * Imprime o valor atual de contador.

  * Incrementa o valor de contador em 1.

- **Repeti√ß√£o**: O loop continua at√© que contador n√£o seja mais menor que 5.

# **‚õî 2.2. Loops Infinitos**

````python
x = 10
while x > 0:
    print(x)
    # Esquecer de decrementar x resulta em loop infinito
````

4 - Convers√£o de Temperatura: Fa√ßa um programa que converta uma temperatura de Celsius para Fahrenheit. Continue pedindo ao usu√°rio para inserir uma nova temperatura em Celsius at√© que ele digite "sair".

````python
# Convers√£o de Temperatura: Celsius para Fahrenheit

def celsius_para_fahrenheit(c):
    return (c * 9/5) + 32

while True:
    entrada = input("Digite a temperatura em Celsius (ou 'sair' para encerrar): ")

    if entrada.lower() == "sair":
        break

    try:
        celsius = float(entrada)
        fahrenheit = celsius_para_fahrenheit(celsius)
        print(f"{celsius}¬∞C √© igual a {fahrenheit:.2f}¬∞F\n")
    except ValueError:
        print("Entrada inv√°lida. Por favor, digite um n√∫mero ou 'sair'.\n")
````

5 - Simulador de Caixa Eletr√¥nico: Crie um programa que simule um caixa eletr√¥nico, que continue pedindo ao usu√°rio para inserir um valor de saque at√© que o saldo da conta seja zero ou negativo.

````python
# Simulador de Caixa Eletr√¥nico

saldo = float(input("Digite o saldo inicial da conta: R$ "))

while saldo > 0:
    print(f"\nSaldo atual: R$ {saldo:.2f}")
    saque = float(input("Digite o valor do saque: R$ "))

    if saque <= 0:
        print("O valor do saque deve ser maior que zero.")
        continue
    elif saque > saldo:
        print("Saldo insuficiente para esse saque.")
        continue

    saldo -= saque
    print(f"Saque de R$ {saque:.2f} realizado com sucesso.")

print("\nSaldo esgotado ou negativo. Encerrando opera√ß√£o.")
````

6 - M√©dia de Notas: Escreva um programa que continue pedindo ao usu√°rio para inserir notas (0 a 10) e calcule a m√©dia dessas notas. O programa deve parar quando o usu√°rio digitar uma nota negativa.

````python
# M√©dia de Notas

notas = []

while True:
    nota = float(input("Digite uma nota entre 0 e 10 (ou uma negativa para sair): "))

    if nota < 0:
        print("Nota negativa detectada. Encerrando...")
        break
    elif nota > 10:
        print("Nota inv√°lida! Digite um valor entre 0 e 10.")
        continue

    notas.append(nota)

# Verifica se alguma nota foi inserida antes de calcular a m√©dia
if notas:
    media = sum(notas) / len(notas)
    print(f"\nForam inseridas {len(notas)} notas.")
    print(f"M√©dia das notas: {media:.2f}")
else:
    print("\nNenhuma nota v√°lida foi inserida.")
````

8- Controle de Estoque: Fa√ßa um controle de estoque com menu de entrada, sa√≠da e exibi√ß√£o.

````python
# Controle de Estoque

estoque = {}

def menu():
    print("\n===== CONTROLE DE ESTOQUE =====")
    print("1 - Entrada de Produto")
    print("2 - Sa√≠da de Produto")
    print("3 - Exibir Estoque")
    print("4 - Sair")
    return input("Escolha uma op√ß√£o: ")

while True:
    opcao = menu()

    if opcao == '1':
        produto = input("Nome do produto para ENTRADA: ").strip()
        quantidade = int(input("Quantidade: "))
        if produto in estoque:
            estoque[produto] += quantidade
        else:
            estoque[produto] = quantidade
        print(f"Adicionado {quantidade} unidades de '{produto}' ao estoque.")

    elif opcao == '2':
        produto = input("Nome do produto para SA√çDA: ").strip()
        if produto not in estoque:
            print("Produto n√£o encontrado no estoque.")
            continue
        quantidade = int(input("Quantidade a remover: "))
        if quantidade > estoque[produto]:
            print(f"Quantidade insuficiente em estoque. Dispon√≠vel: {estoque[produto]}")
        else:
            estoque[produto] -= quantidade
            print(f"Removido {quantidade} unidades de '{produto}'.")
            if estoque[produto] == 0:
                del estoque[produto]  # Remove do dicion√°rio se zerar

    elif opcao == '3':
        if not estoque:
            print("Estoque vazio.")
        else:
            print("\n--- ESTOQUE ATUAL ---")
            for produto, quantidade in estoque.items():
                print(f"{produto}: {quantidade} unidade(s)")

    elif opcao == '4':
        print("Encerrando o programa. At√© mais!")
        break

    else:
        print("Op√ß√£o inv√°lida. Tente novamente.")
````

9 -Simula√ß√£o de Dados de Sensor: Crie um programa que simule a leitura de dados de um sensor e continue capturando dados at√© que um valor fora do intervalo de opera√ß√£o seja detectado (por exemplo, fora de 0 a 100).

````python
import random
import time

# Fun√ß√£o que simula a leitura do sensor
def ler_sensor():
    # Gera um valor aleat√≥rio entre 0 e 100
    return random.uniform(0, 100)

print("=== Simula√ß√£o de Leitura de Dados do Sensor ===")
print("O programa continuar√° lendo os dados at√© que um valor fora do intervalo [0, 100] seja detectado.\n")

while True:
    # Simula a leitura do sensor
    valor_sensor = ler_sensor()

    # Exibe o valor lido
    print(f"Leitura do sensor: {valor_sensor:.2f}")

    # Verifica se o valor est√° fora do intervalo
    if valor_sensor < 0 or valor_sensor > 100:
        print("‚ùå Valor fora do intervalo detectado. Encerrando a simula√ß√£o.")
        break

    # Aguarda 1 segundo antes de ler novamente
    time.sleep(1)
````

10 - Controle de Vota√ß√£o: Fa√ßa um programa que permita cadastrar votos para 3 candidatos. Exibe contagem ao final quando for digitado "fim".

````python
# Controle de Vota√ß√£o

# Dicion√°rio para armazenar os votos dos candidatos
votos = {
    "Candidato 1": 0,
    "Candidato 2": 0,
    "Candidato 3": 0
}

# Fun√ß√£o para exibir o menu
def exibir_menu():
    print("\n=== Menu de Vota√ß√£o ===")
    print("1 - Candidato 1")
    print("2 - Candidato 2")
    print("3 - Candidato 3")
    print("Digite 'fim' para encerrar a vota√ß√£o e mostrar os resultados.")
    return input("Digite o n√∫mero do candidato ou 'fim': ").strip().lower()

print("Bem-vindo ao Sistema de Vota√ß√£o!")
print("Voc√™ pode votar nos candidatos 1, 2 ou 3.")

while True:
    opcao = exibir_menu()

    if opcao == '1':
        votos["Candidato 1"] += 1
        print("Voc√™ votou no Candidato 1.")
    elif opcao == '2':
        votos["Candidato 2"] += 1
        print("Voc√™ votou no Candidato 2.")
    elif opcao == '3':
        votos["Candidato 3"] += 1
        print("Voc√™ votou no Candidato 3.")
    elif opcao == 'fim':
        print("\nFim da vota√ß√£o! Aqui est√£o os resultados:")
        for candidato, qtd_votos in votos.items():
            print(f"{candidato}: {qtd_votos} voto(s)")
        break
    else:
        print("Op√ß√£o inv√°lida. Tente novamente.")
````

11 - Simulador de Caixa Registradora: O usu√°rio digita o pre√ßo e a quantidade de produtos at√© digitar ‚Äúsair‚Äù. O Sistema mostra a quantidade de produtos e o valor final da compra.

````python
# Simulador de Caixa Registradora

total_produtos = 0
valor_total = 0.0

print("=== Simulador de Caixa Registradora ===")
print("Digite 'sair' para finalizar a compra.\n")

while True:
    try:
        preco = input("Digite o pre√ßo do produto (ou 'sair' para finalizar): R$ ")

        if preco.lower() == "sair":
            break

        preco = float(preco)

        if preco < 0:
            print("O pre√ßo n√£o pode ser negativo. Tente novamente.")
            continue

        quantidade = int(input("Digite a quantidade de produtos: "))

        if quantidade < 0:
            print("A quantidade n√£o pode ser negativa. Tente novamente.")
            continue

        total_produtos += quantidade
        valor_total += preco * quantidade

    except ValueError:
        print("Entrada inv√°lida. Por favor, insira um valor num√©rico para o pre√ßo e a quantidade.")
        continue

# Exibe o resumo final
print("\n=== Resumo da Compra ===")
print(f"Quantidade total de produtos: {total_produtos}")
print(f"Valor total da compra: R$ {valor_total:.2f}")
````

12 - Sistema de Vendas com Estoque Reduzido: O Sistema define um estoque inicial de produtos e a cada venda, retira quantidade do estoque. O sistema encerra quando n√£o permite vender mais do que o dispon√≠vel.

````python
# Sistema de Vendas com Estoque Reduzido

# Estoque inicial de produtos (produto: quantidade)
estoque = {
    "Produto A": 50,
    "Produto B": 30,
    "Produto C": 20
}

def exibir_estoque():
    print("\n--- Estoque Atual ---")
    for produto, quantidade in estoque.items():
        print(f"{produto}: {quantidade} unidade(s)")

def realizar_venda(produto, quantidade):
    if produto in estoque and estoque[produto] >= quantidade:
        estoque[produto] -= quantidade
        print(f"Venda realizada! {quantidade} unidade(s) de '{produto}' vendida(s).")
    else:
        print(f"‚ùå N√£o h√° estoque suficiente de '{produto}'. Estoque dispon√≠vel: {estoque.get(produto, 0)}")

def menu():
    print("\n=== Sistema de Vendas ===")
    print("Escolha um produto para vender ou digite 'sair' para encerrar.")
    for i, produto in enumerate(estoque.keys(), 1):
        print(f"{i} - {produto}")
    return input("Digite o n√∫mero do produto ou 'sair': ").strip().lower()

# Loop de vendas
while True:
    exibir_estoque()
    opcao = menu()

    if opcao == "sair":
        print("Encerrando o sistema de vendas.")
        break

    # Verifica se a op√ß√£o √© v√°lida
    if opcao.isdigit() and 1 <= int(opcao) <= len(estoque):
        produto_escolhido = list(estoque.keys())[int(opcao) - 1]
        quantidade = int(input(f"Digite a quantidade de {produto_escolhido} a vender: "))

        # Realiza a venda
        realizar_venda(produto_escolhido, quantidade)
    else:
        print("Op√ß√£o inv√°lida. Tente novamente.")

    # Verifica se todos os produtos t√™m estoque suficiente
    if all(qtd == 0 for qtd in estoque.values()):
        print("\nüö® Estoque de todos os produtos esgotado! O sistema ser√° encerrado.")
        break
````
